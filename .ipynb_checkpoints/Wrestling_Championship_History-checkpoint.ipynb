```python
#Create empty array to append to
goal = []

#Loop through Dictionary key-pair values
for key in Promotion:
        tables = read_html(Promotion[key][0], attrs={'class':'wikitable sortable'})
        tables[0].columns = tables[0].columns.droplevel(0)
        target = tables[0]
        target["Promotion"]=Promotion[key][1]
        target["Title_Name"]=Promotion[key][2]
        goal.append(target)
        new = pd.concat(goal)
        new.to_csv("data3.csv")
```
